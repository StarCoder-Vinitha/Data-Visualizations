```python
import pandas as pd 
import matplotlib as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df
```

### Strip plot

    This is the one kind of scatter plot of categorical data with the help of seaborn.

```python
sns.stripplot(x = 'day',y ='total_bill',data=df)
```

```python
sns.stripplot(x=df['tip'],color = 'green')
```

```python
sns.stripplot(x="day", y="total_bill", data=df)
```

```python
sns.stripplot(x="day", y="tip", data=df)
```

```python
sns.stripplot(x=df['size'],color = 'indigo')
```

```python
sns.stripplot(x=df['total_bill'],color = 'indigo')
```

```python
sns.violinplot(x= 'day',y='tip',data=df)
sns.swarmplot(x='day',y='tip',data = df,color='white')
```
