```python
import pandas as pd 
import matplotlib as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df
```

### joint plot

    seaborn.jointplot () : Draw a plot of two variables with bivariate and univariate graphs. This function provides a convenient interface to the ‘JointGrid’ class, with several canned plot kinds. This is intended to be a fairly lightweight wrapper; if you need more flexibility, you should use :class:’JointGrid’ directly.

```python
sns.jointplot(x = 'total_bill',y='size',data=df,kind='reg',color = 'green')
```

    Here we are knowing that how the size and total_bill was increasing and how the bill was effecting 

```python
sns.jointplot(x = 'tip',y='size',data=df,kind='reg',color = 'blue')
```

```python
sns.jointplot(x = 'size',y='size',data=df,kind='reg',color = 'purple')
```
