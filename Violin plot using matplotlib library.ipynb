```python
import pandas as pd 
import matplotlib as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df
```

### Violin Plot

    seaborn beginner violin plot Violin plots are a great tool to have as an analyst because they allow you to see the underlying distribution of the data while still keeping things clean and simple.

```python
sns.violinplot(x='day',y='total_bill',data=df,hue = 'sex',split=True)
```

```python
sns.violinplot(x='day',y='tip',data=df,split=True,)
```

    The graph showing us which day has recevied most tip = saturday

```python
sns.violinplot(x='day',y='size',data=df,hue = 'sex',split=True)
```
