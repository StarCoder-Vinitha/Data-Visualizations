```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = sns.load_dataset('tips')
df
```

#### Seaborn is a data visualization library based on matplotlib.It provides a highlevel interface for drawing attractive and informative statistical graphics...

### Distribution Plots

#### 1. dist plot

```python
sns.distplot(x = df['total_bill'],color = 'black',bins = 10)
```

#### 2. Joint Plot

```python
sns.jointplot(y = 'tip',x = 'total_bill',data = df,kind = 'reg')
```

```python
sns.jointplot(y = 'tip',x = 'total_bill',data = df,kind = 'hex',color = 'darkcyan')
```

```python
sns.jointplot(y = 'tip',x = 'total_bill',data = df,kind = 'kde',color = 'darkcyan')
```

#### 3. Pair Plot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue = 'sex',palette = 'Blues')
```

#### 4. Rug Plot

```python
sns.rugplot(df['total_bill'])
```
