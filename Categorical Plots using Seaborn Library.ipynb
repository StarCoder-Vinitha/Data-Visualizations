```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = sns.load_dataset('tips')
df
```

### Categorical Plots

#### 1. Bar Plot

```python
sns.barplot(x = 'smoker',y = 'total_bill',data = df,estimator = np.mean)
```

```python
sns.barplot(x = 'smoker',y = 'total_bill',data = df,estimator = np.std)
```

#### 2. Count Plot

```python
sns.countplot(df['smoker'])
```

```python
sns.countplot(df['total_bill'])
```

```python
sns.countplot(df['tip'])
```

#### 3. Box Plot

```python
sns.boxplot(y = df['sex'], x = df['total_bill'])
```

```python
sns.boxplot(y = df['tip'], x = df['total_bill'])
```

```python
sns.boxplot(y = df['total_bill'], x = df['tip'])
```

```python
sns.boxplot(y = df['size'], x = df['tip'])
```

```python
sns.boxplot(y = df['tip'], x = df['size'])
```

#### 4. Violin Plot

```python
sns.violinplot(x = 'sex', y = 'total_bill',hue = 'sex',data = df,split = True)
```

```python
sns.violinplot(x = 'day', y = 'total_bill',hue = 'sex',data = df)
```

```python
sns.violinplot(x = 'day', y = 'tip',hue = 'sex',data = df)
```

```python
sns.violinplot(x = 'day', y = 'tip',hue = 'sex',data = df,split = True)
```

```python
sns.violinplot(x = 'day', y = 'tip',hue = 'sex',data = df,split = False)
```

```python
sns.violinplot(x = 'day', y = 'tip',hue = 'size',data = df)
```

```python
sns.violinplot(x = 'total_bill', y = 'tip',hue = 'sex',data = df)
```

```python
sns.violinplot(y = df['tip'], x = df['total_bill'])
```

```python
sns.violinplot(y = df['total_bill'], x = df['tip'])
```

```python
sns.violinplot(y = df['size'], x = df['tip'])
```

```python
sns.violinplot(y = df['tip'], x = df['total_bill'])
```
